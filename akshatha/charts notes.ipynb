# Power BI Visualizations Explained with Superstore Dataset

## Dataset Columns

| Column | Description |
|------|------|
| Order Date | Date of the order |
| Region | Sales region |
| Category | Product category |
| Sub-Category | Product type |
| Sales | Sales revenue |
| Profit | Profit amount |
| Quantity | Number of products sold |
| Discount | Discount applied |
| Customer Name | Customer |

---

# 1. Bar Chart

## Columns to Choose

Axis → Category  
Values → Sales

## What the Graph Represents

Shows **total sales for each product category**.

Example:

| Category | Sales |
|------|------|
| Furniture | 15000 |
| Technology | 30000 |
| Office Supplies | 20000 |

Higher bar → higher sales.

## Mathematical Representation

Total sales for each category:

$$
Sales_{Category} = \sum_{i=1}^{n} Sales_i
$$

Example:

$$
Sales_{Technology} = 30000
$$

# 3. Line Chart

## Columns to Choose

Axis → Order Date  
Values → Sales

## What the Graph Represents

Shows **sales trend over time**.

Example:

| Month | Sales |
|------|------|
| Jan | 10000 |
| Feb | 12000 |
| Mar | 9000 |

## Mathematical Representation

Sales as a function of time:

$$
Sales = f(t)
$$

Where

$$
t = \text{time (date or month)}
$$

---

# 4. Area Chart

## Columns to Choose

Axis → Order Date  
Values → Sales

## What the Graph Represents

Similar to a line chart but **area under the curve is filled**, representing **cumulative sales growth**.

## Mathematical Representation

Area under the sales curve:

$$
Area = \int_{t_1}^{t_2} Sales(t)\,dt
$$

---

# 5. Pie Chart

## Columns to Choose

Legend → Region  
Values → Sales

## What the Graph Represents

Shows **percentage contribution of each region to total sales**.

Example:

| Region | Sales |
|------|------|
| East | 20000 |
| West | 15000 |
| Central | 10000 |

## Mathematical Representation

$$
Percentage_{Region} =
\frac{Sales_{Region}}{Total\ Sales}
\times 100
$$

Example:

$$
\frac{20000}{45000} \times 100 = 44.4\%
$$

---

# 6. Donut Chart

## Columns to Choose

Legend → Category  
Values → Sales

## What the Graph Represents

Similar to pie chart but with **a hole in the center**.

Shows **share of each category in total sales**.

## Mathematical Representation

$$
Share_{Category} =
\frac{Sales_{Category}}{\sum Sales}
$$

# 9. Scatter Plot

## Columns to Choose

X Axis → Sales  
Y Axis → Profit  
Size → Quantity

## What the Graph Represents

Shows **relationship between two numeric variables**.

Each dot represents **one observation**.

Example:

| Sales | Profit |
|------|------|
| 1000 | 200 |
| 1500 | 100 |

## Mathematical Representation

Relationship between variables:

$$
Profit = f(Sales)
$$

Correlation idea:

$$
r =
\frac{\sum (x_i-\bar{x})(y_i-\bar{y})}
{\sqrt{\sum(x_i-\bar{x})^2 \sum(y_i-\bar{y})^2}}
$$

---

# 10. Tree Map

## Columns to Choose

Group → Category  
Values → Sales

## What the Graph Represents

Shows **hierarchical data using rectangles**.

Large rectangle → larger value.

## Mathematical Representation

Area of rectangle proportional to value:

$$
Area_i \propto Sales_i
$$

---

# 11. Gauge Chart (Speedometer)

## Columns to Choose

Value → Sales  
Target Value → Sales Target

## What the Graph Represents

Shows **progress towards a goal**.

Example:

Target = 50000  
Actual = 35000

## Mathematical Representation

Performance percentage:

$$
Performance =
\frac{Actual}{Target}
\times 100
$$

Example:

$$
\frac{35000}{50000} \times 100 = 70\%
$$
